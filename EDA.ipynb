```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/smartphonecleaned_data.csv')
```

```python
df.sample(5)
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["screen_size"], y=df["price"])
plt.xlabel("Screen Size")
plt.ylabel("Price")
plt.title("Price vs Screen Size")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=df["RAM"], y=df["price"])
plt.xlabel("RAM (GB)")
plt.ylabel("Price")
plt.title("Price vs RAM")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=df["Storage"], y=df["price"])
plt.xlabel("Storage (GB)")
plt.ylabel("Price")
plt.title("Price vs Storage")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["Battery"], y=df["price"])
plt.xlabel("Battery Capacity (mAh)")
plt.ylabel("Price")
plt.title("Price vs Battery")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x=df["processor_speed"], y=df["price"], marker="o")
plt.xlabel("Processor Speed (GHz)")
plt.ylabel("Price")
plt.title("Price vs Processor Speed")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=df["num_rear_cameras"], y=df["price"])
plt.xlabel("Number of Rear Cameras")
plt.ylabel("Price")
plt.title("Price vs Number of Rear Cameras")
plt.show()
```
